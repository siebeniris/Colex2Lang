```python
import pandas as pd

import os
import json

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plt.figure(figsize=(12, 12))
# plt.title("Feature 137B", fontsize=18)

g= sns.relplot(data=df, x="Value", y="Test (F1)", hue="Model", size="Weight",sizes=(100,400) )

g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

```python
# for clics, 

features_clics = ["132A", "133A", "136A", "138A"]

df_list = []
for fid in features_clics:
    file = f"../output/lexicon/clics_{fid}.csv"
    df_feature = pd.read_csv(file)
    df_list.append(df_feature)
    
df_clics = pd.concat(df_list, axis=0)
```

```python
sns.relplot(data=df_clics, x="Value", y="Test (F1)", hue="Model",size="Weight", col="Feature", sizes=(100,400))
# plt.savefig("../output/lexicon/clics.png")
```

```python
sns.relplot(data=df_clics, x="Model", y="Test (F1)", hue="Value",size="Weight", col="Feature", sizes=(100,400))
# plt.savefig("../output/lexicon/clics.png")
```
